# Pandas

Pandas is an open source data analysis library written in python. It leverages the power and speed of numpy to make data analysis and pre processing easy for data scientist.

Pandas has two types of data structures:
1. Series: It ia one dimensional array, consists of single row or column.
2. Dataframe: It is a tabular spreadsheet like structure with columns of different data types.

**Drop Duplicates**

```python
df=pd.DataFrame({"name":["Alfred",'Batman','Alfred'],
                "toy":[np.nan,'Batmobile','Bullwhip'],
                'born':['pd.NaT',pd.Timestamp('1940-04-25'),pd.NaT]})
df
```

```python
df.drop_duplicates(subset=['name'],keep='first')
```

```python
df.drop_duplicates(subset='name',keep='last',inplace=True)
```

```python
df
```

```python
df.reset_index()
```

```python
df.info()
```
