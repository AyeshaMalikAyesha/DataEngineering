# Matplotlib

**Importing libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

**Basics**

```python
x=np.arange(0,100,10)
y=np.arange(0,200,20)

font_dict={
    'family':'serif',
    'color':'darkred',
    'weight':'bold',
     'size':20
    }

plt.plot(x,y)
plt.xlabel('X',fontdict={'color':'green','size':10})
plt.ylabel('Y',fontdict={'color':'purple','size':10})
plt.title('First Graph',fontdict=font_dict)
plt.show()
```

```python
x=np.arange(-10,10,2)
y=np.arange(-10,10,2)

plt.plot(x,y,label='2x',linewidth=2,marker='.',markersize=10,linestyle='--',markeredgecolor='red')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title('Second Graph',color='Red')
plt.show()
```

```python
x1=np.arange(-10,10,2)
y1=np.arange(-10,10,2)

plt.figure(figsize=(8,5))

font={'size':20,
      'fontname':'serif',
      'weight':'normal',
      'color':'red'}

#line number 1
plt.plot(x1[:5],y1[:5],'b^--',label='2x')
plt.plot(x1[4:],y1[4:],'b',label='2x')
plt.title('Second Graph',fontdict=font)

# line number 2
x2=np.arange(0,4,0.5)
plt.plot(x2,x2**2,'r*--',label='x^2')

plt.legend()

# save a graph and dpi=300 is good when saving so graph has high resolution.
plt.savefig('graph.png',dpi=300)

#show plot
plt.show()
```

**Bar Chart**

```python
labels=['Male','Female','Others']
values=[20,30,4]

plt.figure(figsize=(8,5))

bars=plt.bar(labels,values)

patterns=['*','o','/']
for bar in bars:
  bar.set_hatch(patterns.pop())


plt.show()
```

```python
labels=['A','B','C']
values=[20,30,5]

plt.figure(figsize=(8,5))

plt.bar(labels,values,color=['green','yellow','darkred'],width=0.5,edgecolor='black')
plt.title('Students Grades',fontdict={'color':'gray','weight':'bold','size':17})



plt.ylabel('No of students')
plt.xlabel('Grades')

plt.show()
```

**Gas Prices Data**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
gas=pd.read_csv('gas_prices.csv')
gas
```

```python
gas.shape
```

```python
gas.dtypes
```

```python
gas.head()
```

```python
year=gas.Year

plt.plot(year,gas.Australia,'y*--',label='Australia')
plt.plot(year,gas.Canada,'b^--',label='Canada')
plt.plot(year,gas.UK,'r*--',label='UK')

plt.legend()
plt.title('Gas Prices over time in USD',fontdict={'weight':'bold','color':'darkred'})
plt.xlabel('Year')
plt.ylabel('USD')


plt.show()
```

```python
year=gas['Year']

plt.figure(figsize=(7,5))



for country in gas.columns[2:]:
  plt.plot(year,gas[country],label=country)

plt.title('Gas Prices over time in USD')
plt.xlabel('Year')
plt.ylabel('USD')
plt.legend()

plt.savefig('gas_prices_chart.png',dpi=300)
plt.show()
```

**Fifa Data**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fifa=pd.read_csv('fifa_data.csv')
fifa
```

```python
fifa.head(4)
```

**Histogram**

```python
plt.hist(fifa.Overall,bins=50,color='darkred')

plt.xlabel('Skill Level')
plt.ylabel('No. of Players')
plt.title('Distribution of Player Skills')

plt.show()
```

**Pie Chart**

```python
fifa.columns
```

```python
fifa['Preferred Foot'].value_counts()
```

```python
left=fifa.loc[fifa['Preferred Foot']=='Left'].count()[0]
right=fifa.loc[fifa['Preferred Foot']=='Right'].count()[0]

labels=['Left','Right']

plt.pie([left,right],labels=labels,colors=['#7844bd','#a897bd'],autopct='%.2f%%')
plt.title('Foot Preference of fifa Players')

plt.show()
```

```python
fifa.Weight
```

```python
fifa.Weight=[int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]


light=fifa.loc[fifa.Weight<125].count()[0]
medium_light=fifa.loc[(fifa.Weight>=125) & (fifa.Weight<150)].count()[0]
medium=fifa.loc[(fifa.Weight>=150) & (fifa.Weight<175)].count()[0]
medium_heavy=fifa.loc[(fifa.Weight>=175) & (fifa.Weight<200)].count()[0]
heavy=fifa.loc[fifa.Weight>=200].count()[0]

labels=['Under 125','125-150','150-175','175-200','Over 200']

plt.pie([light,medium_light,medium,medium_heavy,heavy],labels=labels,autopct='%.2f%%')
plt.title('Weight of fifa Players')

plt.show()
```

```python
fifa.Weight=[int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]


light=fifa.loc[fifa.Weight<125].count()[0]
medium_light=fifa.loc[(fifa.Weight>=125) & (fifa.Weight<150)].count()[0]
medium=fifa.loc[(fifa.Weight>=150) & (fifa.Weight<175)].count()[0]
medium_heavy=fifa.loc[(fifa.Weight>=175) & (fifa.Weight<200)].count()[0]
heavy=fifa.loc[fifa.Weight>=200].count()[0]

labels=['Under 125','125-150','150-175','175-200','Over 200']
explode=[.1,.1,.1,.1,.1]

plt.pie([light,medium_light,medium,medium_heavy,heavy],labels=labels,autopct='%.2f%%',explode=explode,pctdistance=0.7)
plt.title('Weight of fifa Players in lbs')

plt.show()
```

**Compare Teams**

```python
plt.style.use('default')

plt.figure(figsize=(10,7))

barcelona=fifa.loc[fifa.Club=='FC Barcelona']['Overall']
madrid=fifa.loc[fifa.Club=='Real Madrid']['Overall']
revs=fifa.loc[fifa.Club=='New England Revolution']['Overall']

boxes=plt.boxplot([barcelona,madrid,revs],labels=['barcelona','madrid','revs'],patch_artist=True)

for box in boxes['boxes']:
  box.set(color='red',linewidth=2)
  box.set(facecolor='#e0e0e0')

plt.title('Professional Soccer Team Comparison')
plt.ylabel('FIFA Overall Rating')

plt.show()
```
